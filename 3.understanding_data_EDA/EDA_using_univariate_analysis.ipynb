```python
df = pd.read_csv('tested.csv', index_col='PassengerId')
df.info()
```

# 1. Categorical Data

```python
# CountPlot
# in this data categorical columns are: survived, sex, Embarked
# here we will work with survived and pclass

sns.countplot(df['Pclass'])
df['Survived'].value_counts().plot(kind='bar')
```

```python
# Piechart
df['Survived'].value_counts().plot(kind='pie', autopct='%.2f')
```
