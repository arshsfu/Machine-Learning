# How to frame Machine Learning problem

* First learn the Business problem
* learn the type of problem(in which category of Ml does that problem fall)
* explore the current situation of the problem
* getting data about the problem (carefully examine what features we want and what we dont)
* Metrics to measure (see how much our model is productive based on our findings) 
* want to decide whether we need to go for <b>online learning</b> or <b>Batch learning</b>
* we check our assumptions like is our decided features available or not

<p>usually never start coding directly. Instead jot down everything on paper as per the problem. Explore the problem and then start with coding stuff</p>
