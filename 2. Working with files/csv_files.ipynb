# Opening csv file from an URL

```python
import requests
from io import StringIO
url = "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv"
headers = {"User-Agent":"Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
req = requests.get(url, headers = headers)
data = StringIO(req.text)

pd.read_csv(data)
```

# Sep parameter

```python
pd.read_csv('tested.csv', sep=',')
# default value is , if working with tsv files then enter \t
# if want to add column names if 1st row become column by chance then use parameter called names=[col_val, col_val, ...]
```

# Index_col parameter

```python
pd.read_csv('tested.csv', index_col = 'PassengerId')
```

# Header parameter

```python
pd.read_csv('tested.csv', header = 0)
# if our column name is treated as rows
```

# use_cols paramter

```python
# if we want only some particular columns from dataset 
pd.read_csv('tested.csv', usecols =['Name', 'Sex', 'Age'])
```

# Skiprows/nrows parameter

```python
pd.read_csv('tested.csv', skiprows = [0,1], nrows=100)
# we skip those particular rows in dataframe
# nrows give us no of rows we mentioned 
```

# Handling dates

```python
# All dates are passed as string type so we need to keep those in date format
pd.read_csv('tested.csv', parse_dates=['date'])
# if i have multiple columns one for day other for year , other for month then pass list in parse_dates parameter to combine them into 1 date i.e parse_dates=[[col_name, col_name, col_name]]
```

# convertors

```python
# sometime we need to apply some transformations to some columns of our datset like lets says says RCB shortform whereevr there is Royal challenger bangalore

def rename(name):
    if name == "Royal challenger bangalore":
        return "RCB"
    else:
        return name

# pd.read_csv('tested.csv', converters={'colname':rename})
```

# na_values parameter


```python
# when we have missing values in some rows like - in some places 
# it says what type of values should be considered as NaN values

pd.read_csv('tested.csv', na_values=['male'])
```

# Loading a huge dataset in chunks

```python
# when we dont have enough memory to load huge full dataset then we can load dataset into chunks
dfs = pd.read_csv('tested.csv', chunksize = 100)
```

```python
for chunks in dfs:
    print(chunks.shape)
```
